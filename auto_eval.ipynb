```python
#Make sure all necessary packages are present before running 
import nltk
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('stopwords')
nltk.download('omw-1.4')

from rake_nltk import Rake
import PyDictionary
from nltk.stem import WordNetLemmatizer
```

```python

def break_phrases(list):
    a = []
    for x in list:
        if len(x.split()) == 1:
            a.append(x)
        else:
            a.extend(x.split())
    return a

def lematize(lista):
    w = WordNetLemmatizer()
    a = list(map(w.lemmatize, lista))
    return a

def add_data(data_listing, each_val):
    values_list2 = []
    z = 0
    for x in range(0, len(data_listing)):
        values_list2.append(each_val)
        z += each_val
    values_list2[0] += (100 - sum(values_list2))
    print(values_list2)
    return values_list2

EXPECTED = "A blockchain is a distributed database that is shared among the nodes of a computer network. As a " \
           "database, a blockchain stores information electronically in digital format. Blockchains are best known " \
           "for their crucial role in cryptocurrency systems, such as Bitcoin, for maintaining a secure and " \
           "decentralized record of transactions. The innovation with a blockchain is that it guarantees the fidelity " \
           "and security of a record of data and generates trust without the need for a trusted third party. "
TEST_INP = input("What is Blockchain? \n")

answer_data = TEST_INP
load_answer = EXPECTED

r = Rake()
r.extract_keywords_from_text(load_answer)
data_list = lematize(break_phrases(r.get_ranked_phrases()))
data_list = list(set(data_list))
each_value = int(100 / len(data_list))
values_list = add_data(data_list, each_value)
dict = {data_list[i]: values_list[i] for i in range(0, len(data_list))}

result = 0

print('Existing answer')
print(dict)


max_score = 10

trained_data, test_data = Rake(), Rake()

trained_data.extract_keywords_from_text(load_answer)
test_data.extract_keywords_from_text(answer_data)

test_data_keywords = lematize(break_phrases(test_data.get_ranked_phrases()))

testlist = []
trainlist = []

for x in test_data_keywords:
    testlist.append(x)


z = 0
print('Your answer')
for x in testlist:
    if x in dict.keys():
        print(x)
        result = result + (dict[x] * max_score) / 100
        print(result, dict[x])
    else:
        syn = PyDictionary.PyDictionary().synonym(testlist[z])
        if syn == None:
            continue
        print(syn)
        for j in syn:
            if j in testlist:
                print(dict)
                dict[j] = (dict[x] * max_score) / 100
                result = result + dict[j] * max_score
    z += 1

if result > 10:
    result = 10;
print(f"Score for your answer: {result}")
```
